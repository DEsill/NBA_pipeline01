```python
# delete — from data
import csv
```

```python
with open("dataset/teams_cold.csv", "r") as f:
    reader = csv.reader(f)

    with open("teams_cold.csv", "w", newline="", encoding="utf-8") as f_clean:
        writer = csv.writer(f_clean)

        for row in reader:
            row = [x if x != "—" else "NULL" for x in row]

            writer.writerow(row)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/players_cold.csv')
```

```python
df
```

```python
del df['Unnamed: 0']
```

```python
df.head(50)
```

```python
df["Player"] = df["Player"].str.replace("*", "", regex=False)
```

```python
df.groupby(['Player', 'Year']).get_group(('Greg Anderson', 1991))
```

```python
def single_row(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df['Tm'] == 'TOT']
        row['Tm'] = df.iloc[-1,:]['Tm']
        return row

df = df.groupby(['Player', 'Year']).apply(single_row)
```

```python
df
```

```python
df.index = df.index.droplevel()
```

```python
df.index = df.index.droplevel()
```
